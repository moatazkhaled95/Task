```python
sentences = data.home_text.values
y = data.AI.values
```

```python
sentences_train, sentences_test, y_train, y_test = train_test_split(sentences,y,test_size=0.2, random_state=42)
```

```python
vectorizer = CountVectorizer(min_df=0,lowercase=False)
vectorizer.fit(sentences_train)
```

```python
X_train = vectorizer.transform(sentences_train)
X_test =  vectorizer.transform(sentences_test)
```

```python
X_train
```

```python
input_dim = X_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(100, input_dim=input_dim, activation='relu'))
model.add(Dense(50,  activation='relu'))
model.add(Dense(10,  activation='relu'))






#EarlyStopping(monitor='val_loss')
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
model.summary()
```

```python


history = model.fit(X_train, y_train, epochs = 100, verbose= True, 
                    validation_data=(X_test,y_test), batch_size=32)
```

```python
loss, accuracy = model.evaluate(X_train,y_train, verbose = False)
print("Training Accurracy:{:.4f}".format(accuracy))
loss,accuracy = model.evaluate(X_test,y_test,verbose=False)
print("Testing Accuracy: {:.4f}".format(accuracy))
```
